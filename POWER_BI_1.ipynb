Q.1
Power BI is a business analytics tool developed by Microsoft that helps users connect, analyze, visualize, and share data through interactive dashboards and reports.

 Key Components of Power BI Ecosystem
1. Power BI Desktop
A free application installed on your computer
Used to connect to data sources, clean/transform data, and create reports & visualizations
Best for data modeling and report creation
2. Power BI Service
A cloud-based platform (SaaS)
Used to publish, share, and collaborate on reports and dashboards
Accessible via web browser
Enables real-time data access and teamwork
3. Power BI Mobile
A mobile app available for Android & iOS
Lets users view and interact with dashboards on smartphones/tablets
Supports real-time alerts and notifications
4. Power BI Gateway
A bridge between on-premises data and Power BI Service
Allows secure data refresh and live queries from local databases
Essential when your data is stored locally (not in cloud)

Q.2
Pie Chart vs Donut Chart

Pie Chart

Shows proportions of a whole (100%)
Simple and easy to understand
Best when you have few categories (3–5 max)
Donut Chart
Similar to pie chart but with a hole in the center
Can display additional info (like total value) in the center
Slightly more modern and visually appealing

When to Prefer:

Pie Chart → When you want a quick, simple comparison of parts
Donut Chart → When you want to add extra information in the center or improve design

Example:
Showing market share of brands
Few brands → Pie Chart
Want to show total sales in center → Donut Chart

Bar Chart vs Column Chart

Bar Chart (Horizontal)
Categories on Y-axis, values on X-axis
Best for long category names
Easier to compare many categories

Column Chart (Vertical)
Categories on X-axis, values on Y-axis
Best for time-based data (months, years)
Commonly used for trends

 When to Prefer:
Bar Chart → When you have many categories or long labels
Column Chart → When showing trends over time

 Example:
Comparing sales by 10 different product names (long names) → Bar Chart
Showing monthly sales trend (Jan–Dec) → Column Chart

Q.3

 Star Schema vs Snowflake Schema
Star Schema
Structure: One central fact table connected to multiple dimension tables
Tables are denormalized (less splitting of data)
Simple and fast for reporting

 Significance:

Better performance in Power BI
Easier to understand and build relationships
Snowflake Schema
Structure: Dimension tables are further split into multiple related tables
More normalized (data divided into smaller tables)

Significance:

Reduces data redundancy
More complex and slightly slower in Power BI

 When to Use:
Star Schema → Preferred in Power BI for simplicity & speed
Snowflake Schema → When you need highly structured, normalized data

Primary Key vs Foreign Key (in Power BI)

Primary Key
A column with unique values
Identifies each record in a table

Foreign Key
A column that links to a primary key in another table
Used to create relationships between tables

Significance:

Enables data connection between tables
Helps Power BI perform joins and filtering correctly

Example:

CustomerID in Customer table → Primary Key
CustomerID in Sales table → Foreign Key

Why is Cardinality Important?

Cardinality defines how tables are related based on data uniqueness:

One-to-Many (1:*) → Most common (e.g., one customer → many orders)
Many-to-One (*:1) → Reverse of above
Many-to-Many (:) → More complex, can cause ambiguity

Significance:

Ensures accurate data filtering and calculations
Prevents duplicate or incorrect results
Improves performance and data model clarity

Q.4

Calculated Column vs Measure
Calculated Column
A calculated column is created using DAX and is stored in the table.
It is computed row by row during data load or refresh.
It increases the size of the data model because values are saved.

Use case:
Used when you need a column for filtering, grouping, or creating relationships.

Example:
Profit = Sales − Cost (calculated for each row and stored)

Measure
A measure is also created using DAX but is not stored physically.
It is calculated dynamically at query time based on the current filter context.
It does not increase model size significantly.

Use case:
Used for aggregations such as SUM, COUNT, AVERAGE.

Example:
Total Sales = SUM(Sales[Amount])

Key Differences
Calculated Column: Row-level calculation, stored in memory
Measure: Aggregate-level calculation, computed dynamically
Row Context vs Filter Context
Row Context
Row context refers to calculations that happen for each individual row in a table.
It is automatically applied in calculated columns.

Example:
If a table has Sales and Cost columns:
Profit = Sales − Cost
This is evaluated separately for each row.

Filter Context
Filter context refers to the set of filters applied to the data before calculation.
It affects how measures are evaluated.

Example:
Total Sales = SUM(Sales[Amount])
If a report is filtered to Year = 2024, the measure returns only sales for 2024.

Q.5

Difference Between Report and Dashboard in Power BI
Report
A report is a collection of multiple pages containing visuals such as charts, tables, and graphs.
It is usually created in Power BI Desktop and published to the service.
Reports are interactive and detailed, allowing users to explore data using filters and slicers.
Data in reports can come from multiple datasets.

Use case:
Used for detailed analysis and exploration of data.

Dashboard
A dashboard is a single-page (one canvas) view that displays key visuals.
It is created in the Power BI Service by pinning visuals from reports.
Dashboards are less interactive compared to reports.
A dashboard can combine visuals from multiple reports and datasets.

Use case:
Used for quick monitoring of key metrics (KPIs) in one place.
